# Time Variation

> In air pollution, the variation of a pollutant by time of day and day of week can reveal
useful information concerning the likely sources. For example, road vehicle emissions
tend to follow very regular patterns both on a daily and weekly basis. By contrast
some industrial emissions or pollutants from natural sources (e.g. sea salt aerosol)
may well have very different patterns

Standard libraries to be imported for usage

```python
import pandas as pd
from vayu import timeVariation
```

Reading the data and using the function from the library.

```python
df = pd.read_csv("../data/mydata.csv")
timeVariation(df, ['pm10','no2','pm25', 'so2'])
```
